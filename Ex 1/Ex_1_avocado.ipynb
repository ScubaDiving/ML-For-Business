```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("avocado.csv")

# Display first 5 lines
print(df.head())

```

```python
# Display last 10 lines
print(df.tail(10))
```

```python
# Number of rows and columns
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")

# Alternative method and more information
print(df.info())
```

```python
# Columns with missing values
print(df.isnull().sum())
```

```python
### Data manipulation

# Mean value of one column
column_name = "XLarge Bags"  
mean_value = df[column_name].mean()
print(f"Mean of {column_name}: {mean_value}")

```

```python
# Create a new dataframe with 3 columns and 2000 rows
small_df = df[list(df)[0:3]].sample(2000)

# Group by categorical attribute and calculate mean of a numerical attribute
group_by_column = "type"  
numeric_column = "XLarge Bags"  
grouped_data = df.groupby(group_by_column)[numeric_column].mean()
print(grouped_data)
```

```python
# Count entries for each category in the above calculation
count_entries = df.groupby(group_by_column)[numeric_column].count()
print(count_entries)
```

### Data Visualization

```python
# Histogram
column_name = "AveragePrice" 
plt.hist(df[column_name], bins=20)
plt.xlabel("Average Price")
plt.ylabel("Frequency")
plt.title(f"Histogram of {column_name}")
plt.show()

# Scatter plot
x_column = "AveragePrice"
y_column = "Total Bags"  
plt.scatter(df[x_column], df[y_column], alpha=0.5)
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.title(f"Scatter plot of {x_column} vs. {y_column}")
plt.show()
```
